# **Preprocessing**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
db_scala = db_sel[['Engine Size','Consumption Comb', 'Cylinders', 'Emissions']]
db_scala = pd.DataFrame(scaler.fit_transform(db_scala), columns = db_scala.columns)
```

```python
from sklearn.model_selection import train_test_split

X = db_scala[['Engine Size','Consumption Comb', 'Cylinders']]
y = db_scala['Emissions']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 1)
```

# **Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
lr = LinearRegression().fit(X_train, y_train)

y_pred = lr.predict(X_test)

lr_score = r2_score(y_test, y_pred)
lr_rmse = mean_squared_error(y_test, y_pred, squared = False)
print("Coefficients: ", lr.coef_)
print("Intercept: ", lr.intercept_)
print("R2 Score : ", lr_score)
print("RMSE : ", lr_rmse)
```

Dari scatterplot di cell [14], grafik menunjukkan hubungan linier. Akan tetapi, bahan bakar jenis diesel dan etanol terlihat terpisah, jadi saya mencoba menggunakan Decision Tree and Random Forest.

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor().fit(X_train, y_train)

y_pred = dtr.predict(X_test)

dtr_score = r2_score(y_test, y_pred)
dtr_rmse = mean_squared_error(y_test, y_pred, squared = False)
print("R2 Score : ", dtr_score)
print("RMSE : ", dtr_rmse)
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor().fit(X_train, y_train)

y_pred = rfr.predict(X_test)

rfr_score = r2_score(y_test, y_pred)
rfr_rmse = mean_squared_error(y_test, y_pred, squared = False)
print("R2 Score : ", rfr_score)
print("RMSE : ", rfr_rmse)
```

```python
log = pd.DataFrame({
    'Name' : ['Linear Regression', 'Decision Tree', 'RandomForestRegressor'],
    'R2 Score' : [lr_score, dtr_score, rfr_score],
    'RMSE' : [lr_rmse, dtr_rmse, rfr_rmse]
})
log
```

```python
plt.title('R2 Score')
plt.bar(log['Name'], log['R2 Score'], color = ['r','g','b','y'])
plt.xlabel('Model')
plt.ylabel('r^2 score')
plt.ylim(0, 1)
```

```python
#RMSE adalah Root Mean Square Error
plt.title('rmse')
plt.bar(log['Name'], log['RMSE'], color = ['r','g','b','y'])
plt.xlabel('Model')
plt.ylabel('RMSE')
```

Hasilnya, Random Forest Regressor memiliki performa terbaik dibandingkan dengan R2 Score : 0.978925 RMSE : 0.019842
